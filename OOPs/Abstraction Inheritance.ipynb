# Inheritance

```python
class test: #PARENT CLASS
    def __init__(self, a, b, c, d):
        self.__a = a
        self.b = b
        self.c = c 
        self.d = d
    def test_custom(self, v):
        return v-self.__a  #when replaced by _ _a it works
    def __str__(self):
        return"this is my test code for abstraction"
```

```python
class test1(test):   #CHILD CLASS (PARENT CLASS)
    def __init__(self, j, *args):
        super(test1,self).__init__(*args) #super of test1 -> test
        self.j = j

m = test1(1,2,3,4,5)
#   test1(j,test(*args)) : --------->  (j , a , b, c , d)
m.b
```

```python
m.j
```

```python
m.test_custom(8)   # CHILD CLASS is able to call a method from PARENT CLASS
```

# Multiple Inheritance



```python
# Python Program to depict multiple inheritance
# when method is overridden in both classes

class test1:
    def m(self):
        print("In Class test1")
        
class test2:
    def m(self):
        print("In Class test2")
    
class test3(test1,test2):
    def m(self):
        test1.m(self)
        test2.m(self)
        print("In Class Test3")
    
obj = test3()
obj.m()
```

```python
test1.mro() #Method Resolution
```

```python
class Employees(): 
 
   def Name(self): 
       print("Employee Name: Khush")
    
class salary():
    def Salary(self):
       print("Salary: 10000")
    
class Designation(Employees, salary):
    def desig(self):
       print("Designation: Test Engineer")
 
call = Designation()

# All the three methods are inherited by the Designation Object
call.Name()
call.Salary()
call.desig()
```

```python
#Class Variables are defined inside class and can be utilized through inheritance.
# e.g. : self.name , self.yearofest

class zf:
    website = "www.zf.com"
    name = "ZF India Pvt Ltd"
    def contact_details(self):
        print("Call us at 1800-201-047")

class diwid(zf):
    def __init__(self):
        self.yearofest = 2017
        
    def est_details(self):
        print('{0} Company was established in {1}'
              .format(self.name , self.yearofest))

        
        
        
class os:
    multi_task = True
    os_name = "Win+10"
    name = "Microsoft"
    def contact_details(self):
        print("Call us at 1800-201-047")
        
class windows(os,zf):   #inheritance from OS and ZF Classes 
                        #(Prioritises OS over ZF)
    def __init__(self):
        if self.multi_task:
            print("OS Supports MultiTask")
        print("Name :{0}".format(self.name))
        
        
div = diwid()
div.est_details()

print("\n\n")

win = windows()
windows.mro() # method resolution order.(Precedence Order)
```

# 3 Level of Inheritance

```python
class zf:
    num_of_emp = 3000

class diwid(zf):
    div_type = "Innovation"

class AIML(diwid):
    def __init__(self):
        self.company = 'ZF'
        print("zf: {0} \ndiwid: {1}\nAIML: {2}"
              .format(self.num_of_emp , self.div_type , self.company))
        
VJ = AIML()

# Class AIML access all the variables from the Parent Classes
```

# Encapsulation
Encapsulation is defined as the wrapping up of data under a single unit. It is the mechanism that binds together code and the data it manipulates. Another way to think about encapsulation is, that it is a protective shield that prevents the data from being accessed by the code outside this shield.
