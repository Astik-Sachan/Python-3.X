```python
class car: #a below is same as self
    def __init__(a , m , y , n='unknown'): #def __init__(mileage , year):
        a.mileage = m
        a.year = y
        a.name = n
        
    def age(self,current_year):
        return current_year - self.year
    
    def mileage_print(self):
        print('Mileage of this car is  : ',self.mileage)
        return self.mileage
    
    def __str__(self): #Print Class Detail
        return "This is my CAR Class it contains mileage, year , name\n and few methods like mileage_print()"
    
        
        
# a : Pointer to the class
# m,y,n : Mileage - class variables
# self : is not a keyword its just another variable.

n = car(10,2000)
```

```python
class student:
    def __init__(self, name, roll, joiningdate, current_trade):
        self.name = name
        self.roll = roll
        self.joiningdate = joiningdate
        self.current_trade = current_trade
        
    def crt_trade(self):
        print("Current topic discussed in my class is : ",self.current_trade)
        
    def str_rollno(self):   
        if type(self.roll) == str:
            print('Do Nothing')
        else:
            return str(self.roll)

            
    def duration(self,current_date):
        print("Duration of student in my class is ", current_date - self.joiningdate)
        
    def __str__(self):
        return "This is an object of Student CLASS"
    
    def name_parse(self):
        if type(self.name)==list:
            for i in self.name :
                print("name of the student : ",i)
        else:
            print("Provided name is not in form of list : ",self.name)
```
